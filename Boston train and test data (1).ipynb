```python
from sklearn import datasets
import pandas as pd
```

```python
boston=datasets.load_boston()
boston
```

```python
X=boston.data
Y=boston.target

type(X)
type(Y)
X.shape
X
```

```python
df=pd.DataFrame(X)
df.columns=boston.feature_names
df["Age_Age"]=df.AGE**2
df["CRIM_CRIM"]=df.CRIM**2
df["INDUS_INDUS"]=df.INDUS**2
df["CHAS_CHAS"]=df.CHAS**2
X2=df.values
df
X2.shape
df
```

```python
pd.DataFrame(Y)
```

```python
from sklearn import model_selection
X_train,X_test,Y_train,Y_test=model_selection.train_test_split(X,Y,random_state=0)
X2_train,X2_test,Y2_train,Y2_test=model_selection.train_test_split(X2,Y,random_state=0)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
print(X2_train.shape)
print(X2_test.shape)
print(Y2_train.shape)
print(Y2_test.shape)

```

```python
from sklearn.linear_model import LinearRegression
algo1=LinearRegression()
algo2=LinearRegression()
```

```python
algo1.fit(X_train, Y_train)
algo2.fit(X2_train, Y2_train)
```

```python
Y_pred = algo1.predict(X_test)
train_score = algo1.score(X_train, Y_train)
test_score = algo1.score(X_test, Y_test)
print("Train Score: ", train_score)
print("Test Score: ", test_score)


train2_score = algo2.score(X2_train, Y2_train)
test2_score = algo2.score(X2_test, Y2_test)
print("Train2 Score: ", train2_score)
print("Test2 Score: ", test2_score)
```

```python

# compare Y _pred and Y_test
import matplotlib.pyplot as plt
plt.scatter(Y_test,Y_pred)
plt.axis([0,40,0,40])
plt.show()
```
